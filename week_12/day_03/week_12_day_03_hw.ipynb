# Task 1

```python
card_base = pd.read_csv('data/CardBase.csv')
customer_base = pd.read_csv('data/CustomerBase.csv')
fraud_base = pd.read_csv('data/FraudBase.csv')
transaction_base = pd.read_csv('data/TransactionBase.csv')
```

```python
def make_cols_snake_case(df) :
    # make list of new clean col names
    clean_cols = [x.lower().replace(" ","_").strip() for x in df.columns.tolist()]
    # overwrite existing col names with clean ones
    df.columns = clean_cols
    return(df)

make_cols_snake_case(card_base)
make_cols_snake_case(customer_base)
make_cols_snake_case(fraud_base)
make_cols_snake_case(transaction_base)
```
